***To Remove the outliers in the data we would be adding their mean in_place of null values***

```python
import pandas as pd

columns_to_fill = Year2008.loc[:, 'All Categories - Classes I-V - Boys':'Scheduled Tribe - Classes I-X - Total']

means = columns_to_fill.mean(axis=1).round(2)

for index, row in columns_to_fill.iterrows():
    Year2008.loc[index, 'All Categories - Classes I-V - Boys':'Scheduled Tribe - Classes I-X - Total'] = row.fillna(means[index])

print(Year2008)

Year2008.to_csv('ne1.csv', index=False)
```

# 2. What were the change in dropout ratio for class I-X for SC, ST and All categories over the four years?

# 3. What was the change in dropout ratio for class I-X Girls for 2008 and 2009?

```python
import matplotlib.pyplot as plt

states = merge['States / Union Territories']
boys_performance = merge['All Categories - Classes I-X - Girls_2008']  
girls_performance = merge['All Categories - Classes I-X - Girls_2009'] 

bar_width = 0.4

x = range(len(states))

plt.figure(figsize=(12, 6))

plt.bar(x, boys_performance, width=bar_width, label='2008', align='center', alpha=0.7)

plt.bar([i + bar_width for i in x], girls_performance, width=bar_width, label='2009', align='center', alpha=0.7)

plt.xlabel("State/Union Territory")
plt.ylabel("No. of % of girls")
plt.title("Dropout rates by State/Union Territory (2008 and 2009) - Of Girls")
plt.xticks([i + bar_width / 2 for i in x], states, rotation=90)
plt.legend()
plt.show()
```

# 4. What was the change in dropout ratio for class I-V Girls for 2010 and 2011?

```python
import matplotlib.pyplot as plt

states = merge['States / Union Territories']
boys_performance = merge['All Categories - Classes I-X - Girls_2010'] 
girls_performance = merge['All Categories - Classes I-X - Girls_2011']

bar_width = 0.4

x = range(len(states))

plt.figure(figsize=(12, 6))

plt.bar(x, boys_performance, width=bar_width, label='2010', align='center', alpha=0.7)

plt.bar([i + bar_width for i in x], girls_performance, width=bar_width, label='2011', align='center', alpha=0.7)

plt.xlabel("State/Union Territory")
plt.ylabel("Educational Performance")
plt.title("Dropout rates by State/Union Territory (2010 and 2011) - of Girls")
plt.xticks([i + bar_width / 2 for i in x], states, rotation=90)
plt.legend()
plt.show()
```

1. Heat Map:
-> In the heat map, in the correllation between "All Categories Boys and Girls" there was a dip in dropout rate in 2011.
-> Correlation between "SC category Boys and Girls", there was decrement in dropout rate until 2011 there was a slight spike.
-> For the category of "ST Boys and Girls", in the years 2009 and 2011 their were lowest rates.

2. Box Plot:
Box plot was implemented to check at what rate the outliers were present in the given dataset. After viewing the visualisation there were high discrepency in the way the data was given.

3. Bar Graph:
-> 2008-09; We took the column of "All category Girls of I-V". And there we observed, between the two years Puducherry and U.P had the lowest girl dropout rate in the country. Overall also there was a decrease in rates in many states.
-> 2010-11; Overall we depicted an increment in dropout rates for the most of the states in which Punjab showed a huge rise of 25-27%.

4. Pair Plot:
In the above performed pair plot, we can see that most of the graphs had scattered dots, not linear. This indicates a lack of strong linear correlation which means that for the given column value if there is an increase im the near future there would be a slight chance of increase for the other column values.

5. Pie Chart:
-> I-V Year-wise and Caste-wise; We saw a zig - zag level of dropout rates for the "ST category" students which was not uniform.
-> I-VIII Year-wise and Caste-wise; There was a decrement in "SC category" dropout rates after that there was an increment by 2-3%.
-> I-X Year-wise and Caste-wise; We saw a constant rate i.e., there was neither decrement nor increment in data for "All categories".

6. -> Andhra Pradesh, Arunachal Pradesh, Assam, Bihar, Chhattisgarh were the lowest states showing the dropouts.
-> After analyzing the data we found that Delhi, Lakshadweep and Puducherry were among the top states which showed highest dropout rates.*
